```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
data = pd.read_csv("insurance.csv")

# Print the data
print(data.head())

# Plot scatter plot
plt.figure(figsize=(10,6))
plt.scatter(data['age'], data['charges'], alpha=0.5)
plt.title('Scatter plot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()

# Prepare the data for modeling
X = data['age'].values.reshape(-1,1)
y = data['charges'].values

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot scatter plot with best-fit line
plt.figure(figsize=(10,6))
plt.scatter(data['age'], data['charges'], alpha=0.5, label='Actual Data')
plt.plot(data['age'], predictions, color='red', label='Best Fit Line')
plt.title('Age vs Charges with Best Fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.grid(True)
plt.show()
```
